##### The plan is to make a for on for loop using the url the outer loop is TTT and the inner loop is LOS 
##### TTT = the gap between the day of search with the day of fly 
##### LOS = time of vac

```python
from datetime import datetime, timedelta
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
```

```python
first_part_url = "https://www.il.kayak.com/flights/"
last_part_url = "?ucs=1tehueu&sort=bestflight_a"
```

```python
dep_city = [ 'LON' ]
back_city = [ 'PAR' ]
```

```python
TTT = [(datetime.today() + timedelta(days=i)).strftime('%Y-%m-%d') for i in range(1, 31)]
```

```python
LOS = [(datetime.today() + timedelta(days=i)).strftime('%Y-%m-%d') for i in range(1, 6)]
LOS
```

```python
all_urls = []

# Create the URLs
for dep in dep_city:
    for back in back_city:
        # Skip if departure city and return city are the same
        if dep == back:
            continue
        for t in TTT:
            for l in LOS:
                # Notice we use "/{t}/{l}" instead of "-{t}-{l}"
                url = f"{first_part_url}{dep}-{back}/{t}/{l}{last_part_url}"
                all_urls.append(url)

print(f"Total URLs generated: {len(all_urls)}")
print("Sample URLs:")
for u in all_urls[:]:  # Show just the first 10
    print(u)
```

```python
driver = webdriver.Chrome()
driver.get('https://www.il.kayak.com/flights/LON-PAR/2025-02-27/2025-02-27?ucs=1tehueu&sort=bestflight_a')
```

```python
#Catch each flight and return how much flight in page
elements = driver.find_elements(By.XPATH, './/*[@class="Fxw9-result-item-container"]')

# Get the length of the elements list
length = len(elements)

print (length)
```

```python
#click show more until we have 100 difrrent flights
elements = driver.find_elements(By.XPATH, './/*[@class="Fxw9-result-item-container"]')
length = len(elements)

# Keep clicking the 'Show More' button until 100 results are present
while length < 100:
    # Locate 'Show More' button
    load_more_buttons = driver.find_elements(By.XPATH, '//*[@class="ULvh-button show-more-button"]')
    
    if load_more_buttons:  # Ensure at least one button exists
        load_more_buttons[0].click()  # Click the first button
        time.sleep(2)  # Wait for new results to load
        elements = driver.find_elements(By.XPATH, './/*[@class="Fxw9-result-item-container"]')
        length = len(elements)  # Update the length of elements
```
